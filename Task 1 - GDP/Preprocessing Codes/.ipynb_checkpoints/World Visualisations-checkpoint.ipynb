```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Final.csv'
data = pd.read_csv(file_path)
```

```python
# Ensure 'Year' column exists and is clean
if 'Year' in data.columns:
    data = data.dropna(subset=['Year'])  # Remove rows with missing 'Year'
    data['Year'] = data['Year'].astype(int)  # Ensure 'Year' is an integer
    data = data.set_index('Year')
else:
    print("Error: 'Year' column not found in the dataset.")
```

```python
# Remove columns with completely missing data
data = data.dropna(axis=1, how='all')  # Drop columns with all NaN values

# Convert all values to numeric, handling missing data
data = data.apply(pd.to_numeric, errors='coerce')

# Replace NaN values with the mean of the respective columns (countries)
data = data.apply(lambda x: x.fillna(x.mean()), axis=0)

# Confirm the data cleaning process
print("Data cleaned successfully. Sample data:")
print(data.head())
```

```python
# Visualization 1: Aggregate GDP of all countries over time
world_gdp = data.sum(axis=1)
plt.figure(figsize=(14, 7))
plt.plot(world_gdp.index, world_gdp, label='World GDP', marker='o')
plt.title('World GDP Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (in billions USD)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Visualization 2: Top 10 countries contributing to global GDP in the most recent year
latest_year = data.index.max()
top_countries = data.loc[latest_year].sort_values(ascending=False).head(10)
plt.figure(figsize=(14, 7))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title(f'Top 10 Countries by GDP in {latest_year}', fontsize=14)
plt.xlabel('GDP (in billions USD)', fontsize=12)
plt.ylabel('country', fontsize=12)
plt.show()
```

```python
# Visualization 3: Annual percentage growth of world GDP
world_gdp_growth = world_gdp.pct_change() * 100
plt.figure(figsize=(14, 7))
plt.plot(world_gdp_growth.index, world_gdp_growth, label='World GDP Growth Rate', marker='o')
plt.axhline(0, color='red', linestyle='--', label='No Growth')
plt.title('World GDP Growth Rate Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Rate (%)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Ensure the latest year is 2024
latest_year = 2024

# Ensure that USA and Japan are always in the top 10
top_countries = data.loc[latest_year].sort_values(ascending=False).head(10).index.tolist()

# Add USA and Japan to the top countries if they are not already included
if 'United States' not in top_countries:
    top_countries.append('United States')
if 'Japan' not in top_countries:
    top_countries.append('Japan')

# Subset the data to include only the top 10 countries (including USA and Japan)
top_countries_data = data.loc[latest_year, top_countries]

# Visualization 1: Bar chart for the top 10 countries GDP distribution
plt.figure(figsize=(14, 7))
top_countries_data.sort_values(ascending=False).plot(kind='bar', color='teal', alpha=0.7)
plt.title(f'GDP Distribution Across Top 10 Countries in {latest_year}', fontsize=14)
plt.xlabel('country', fontsize=12)
plt.ylabel('GDP (in billions USD)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Visualization 2: Pie chart for the GDP share of the top 10 countries
plt.figure(figsize=(10, 10))
top_countries_data.sort_values(ascending=False).plot(
    kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title(f'GDP Share of Top 10 Countries in {latest_year}', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'world_bank_development_indicators.csv'
data = pd.read_csv(file_path)

# Filter data for USA and Japan
countries_of_interest = ['United States', 'Japan']
filtered_data = data[data['country'].isin(countries_of_interest)]

# Select relevant columns for analysis
columns_of_interest = [
    'country', 'date', 'trade_in_services%', 'control_of_corruption_estimate',
    'access_to_electricity%', 'population', 'birth_rate', 'death_rate',
    'life_expectancy_at_birth', 'agricultural_land%', 'multidimensional_poverty_headcount_ratio%'
]
filtered_data = filtered_data[columns_of_interest]

# Ensure the date column is in datetime format for better plotting
filtered_data['date'] = pd.to_datetime(filtered_data['date'])

# Set up subplots for visualizations
fig, axes = plt.subplots(4, 1, figsize=(14, 20), sharex=True)

# Population trends over time
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    axes[0].plot(country_data['date'], country_data['population'], label=country)

axes[0].set_title('Population Trends (USA vs. Japan)', fontsize=14)
axes[0].set_ylabel('Population', fontsize=12)
axes[0].legend()
axes[0].grid()

# Life expectancy trends over time
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    axes[1].plot(country_data['date'], country_data['life_expectancy_at_birth'], label=country)

axes[1].set_title('Life Expectancy Trends (USA vs. Japan)', fontsize=14)
axes[1].set_ylabel('Life Expectancy at Birth', fontsize=12)
axes[1].legend()
axes[1].grid()

# Trade in Services trends over time
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    axes[2].plot(country_data['date'], country_data['trade_in_services%'], label=country)

axes[2].set_title('Trade in Services (% of GDP) Trends (USA vs. Japan)', fontsize=14)
axes[2].set_ylabel('Trade in Services (%)', fontsize=12)
axes[2].legend()
axes[2].grid()

# Agricultural Land trends over time
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    axes[3].plot(country_data['date'], country_data['agricultural_land%'], label=country)

axes[3].set_title('Agricultural Land (% of Total) Trends (USA vs. Japan)', fontsize=14)
axes[3].set_ylabel('Agricultural Land (%)', fontsize=12)
axes[3].set_xlabel('Year', fontsize=12)
axes[3].legend()
axes[3].grid()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Select numeric columns for correlation analysis
numeric_data = filtered_data.drop(['country', 'date'], axis=1).dropna()

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Factors Affecting GDP')
plt.show()
```

```python
# Reshape the data for plotting
trade_agriculture_data = filtered_data[['country', 'date', 'trade_in_services%', 'agricultural_land%']].dropna()
pivoted_data = trade_agriculture_data.pivot(index='date', columns='country', values=['trade_in_services%', 'agricultural_land%'])

# Plot stacked bar chart
pivoted_data.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Trade in Services and Agricultural Land (% of GDP)')
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.legend(title='country and Factor')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    plt.scatter(country_data['population'], country_data['trade_in_services%'], label=country)

plt.title('Population vs. Trade in Services (% of GDP)')
plt.xlabel('Population')
plt.ylabel('Trade in Services (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Filter data for the latest year
latest_year_data = filtered_data[filtered_data['date'] == filtered_data['date'].max()]

for country in countries_of_interest:
    country_data = latest_year_data[latest_year_data['country'] == country]
    
    if country_data['trade_in_services%'].isna().any() or country_data['agricultural_land%'].isna().any():
        print(f"Missing data for {country}: Trade={country_data['trade_in_services%'].values[0]}, Agri={country_data['agricultural_land%'].values[0]}")
        continue  # Skip this country if data is missing

    labels = ['Trade in Services (%)', 'Agricultural Land (%)']
    sizes = [country_data['trade_in_services%'].values[0], country_data['agricultural_land%'].values[0]]

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen'])
    plt.title(f'Contribution of Agriculture and Trade in {country} (Latest Year)')
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
for country in countries_of_interest:
    country_data = filtered_data[filtered_data['country'] == country]
    plt.scatter(
        country_data['population'], 
        country_data['trade_in_services%'], 
        s=country_data['life_expectancy_at_birth'] * 5,  # Bubble size based on life expectancy
        alpha=0.6, label=country
    )

plt.title('Population vs. Trade in Services (% of GDP) with Life Expectancy as Bubble Size')
plt.xlabel('Population')
plt.ylabel('Trade in Services (%)')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "world_bank_development_indicators_with_gdp_per_capita.csv"
data = pd.read_csv(file_path)

# Filter data for Japan and USA
filtered_data = data[(data['country'] == 'Japan') | (data['country'] == 'United States')]

# Fill missing values with the mean for numeric columns
filtered_data.fillna(filtered_data.mean(numeric_only=True), inplace=True)

# Convert date column to datetime without specifying the format
filtered_data['date'] = pd.to_datetime(filtered_data['date'], errors='coerce').dt.year

# Bubble chart: GDP vs Trade, bubble size representing CO2 emissions
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=filtered_data,
    x="GDP_current_US",
    y="trade_in_services%",
    hue="country",
    size="CO2_emisions",
    sizes=(50, 500),
    alpha=0.7,
    palette="muted"
)
plt.title("Bubble Chart: GDP vs Trade in Services with CO2 Emissions as Bubble Size")
plt.xlabel("GDP (current US$)")
plt.ylabel("Trade in Services (% of GDP)")
plt.legend(title="Country", loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

# Area plot: Renewable Energy Consumption over time
plt.figure(figsize=(14, 8))
sns.lineplot(
    data=filtered_data,
    x="date",
    y="renewvable_energy_consumption%",
    hue="country",
    marker="o",
    linewidth=2
)
plt.fill_between(
    filtered_data["date"],
    filtered_data["renewvable_energy_consumption%"],
    alpha=0.2
)
plt.title("Area Plot: Renewable Energy Consumption Over Time")
plt.xlabel("Year")
plt.ylabel("Renewable Energy Consumption (%)")
plt.legend(title="Country")
plt.tight_layout()
plt.show()

# Stacked bar chart: Contribution of tax revenue, expense, and central government debt
stacked_data = filtered_data.pivot_table(
    index="date",
    columns="country",
    values=["tax_revenue%", "expense%", "central_goverment_debt%"],
    aggfunc="mean"
)

stacked_data.plot(kind="bar", stacked=True, figsize=(14, 8), colormap="coolwarm")
plt.title("Stacked Bar Chart: Tax Revenue, Expense, and Central Government Debt")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend(title="Indicators", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# Population Density vs Internet Usage: Bubble chart
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=filtered_data,
    x="population_density",
    y="individuals_using_internet%",
    hue="country",
    size="population",
    sizes=(50, 500),
    alpha=0.7,
    palette="Set2"
)
plt.title("Bubble Chart: Population Density vs Internet Usage with Population as Bubble Size")
plt.xlabel("Population Density (people per sq. km)")
plt.ylabel("Individuals Using Internet (%)")
plt.legend(title="Country", loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
