```python
import pandas as pd

# Assuming 'filtered_gwp_ar2' is the DataFrame

# List of columns representing the years 2005 to 2015
year_columns = [str(year) for year in range(2005, 2016)]

# Dictionary to store results for each category
category_sums = {}

# Calculate the summation for each category in 'Level 1'
for category in filtered_gwp_ar2['Level 1'].unique():
    # Filter rows for the current category
    category_df = filtered_gwp_ar2[filtered_gwp_ar2['Level 1'] == category]
    
    # Sum emissions for each year
    category_sums[category] = category_df[year_columns].sum()

# Convert the dictionary to a DataFrame for easier display and export
category_sums_df = pd.DataFrame(category_sums, index=year_columns).transpose()

# Add a row for the final summation for all categories
category_sums_df.loc['Total'] = category_sums_df.sum()

# Display the results
print(category_sums_df)

# Save the results to a CSV file
category_sums_df.to_csv('yearly_emissions_by_category_and_total.csv')
print("Yearly emissions by category and totals saved to 'yearly_emissions_by_category_and_total.csv'")
```

```python
# Remove the 'Total' row for category-based plotting
category_df = yearly_emissions_by_category_and_total[yearly_emissions_by_category_and_total['Category'] != 'Total']

# Extract the "Total" row for line plotting
total_emissions = yearly_emissions_by_category_and_total[yearly_emissions_by_category_and_total['Category'] == 'Total']

# List of years
years = [str(year) for year in range(2005, 2016)]

# Plotting
fig, ax = plt.subplots(figsize=(15, 8))

# Bar width
bar_width = 0.2
x_indexes = range(len(years))

# Iterate through each category and plot its bars
for i, category in enumerate(category_df['Category']):
    ax.bar(
        [x + i * bar_width for x in x_indexes],  # Offset bars for each category
        category_df.loc[category_df['Category'] == category, years].values.flatten(),
        width=bar_width,
        label=category
    )

# Plot the total emissions as a line
ax.plot(
    [x + 1.5 * bar_width for x in x_indexes],  # Align with the center of grouped bars
    total_emissions[years].values.flatten(),
    color='black',
    marker='o',
    linestyle='--',
    linewidth=2,
    label='Total'
)

# Customizations
ax.set_title("Yearly Emissions by Category", fontsize=16)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Emissions (in tonnes)", fontsize=14)
ax.set_xticks([x + 1.5 * bar_width for x in x_indexes])  # Center labels
ax.set_xticklabels(years, rotation=45)
ax.legend(title="Category", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for trigonometric functions

def plot_greenhouse_gas_chart():
    # Data for the chart
    gases = ["CO₂", "CH₄", "N₂O", "Others"]
    percentages = [80, 15, 3, 2]

    # Colors for each gas
    colors = ["gray", "#3C8031", "#92C5E0", "red"]

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    wedges, texts = ax.pie(
        percentages,
        labels=None,  # Remove default labels
        colors=colors,
        startangle=90,
        counterclock=False,
        wedgeprops={"width": 0.3, "edgecolor": "white"},
    )

    # Add percentages inside or outside the chart
    for wedge, percent in zip(wedges, percentages):
        angle = (wedge.theta2 + wedge.theta1) / 2  # Find the center angle of the wedge

        # Adjust the factor here to control text placement (0.6 moves it closer, 0.8 moves it outward)
        radius_factor = 0.8
        x = radius_factor * np.cos(np.radians(angle))
        y = radius_factor * np.sin(np.radians(angle))

        ax.text(x, y, f"{percent}%", ha="center", va="center", fontsize=10, color="black")

    # Add legend outside the chart
    ax.legend(
        labels=gases,
        loc="center left",
        bbox_to_anchor=(1, 0.5),
        fontsize=10,
        title="Greenhouse Gases"
    )

    # Title with reduced padding
    plt.title("Percentage Share of Greenhouse Gases (2010)", fontsize=12, pad=10)

    # Display the plot
    plt.show()

# Plot the chart
plot_greenhouse_gas_chart()
```

```python
import pandas as pd

# Load the data from each CSV file
energy = pd.read_csv('top_states_2010_energy.csv')
ippu = pd.read_csv('top_states_2010_ippu.csv')
waste = pd.read_csv('top_states_2010_waste.csv')
afolu = pd.read_csv('top_states_2010_afolu.csv')

# Merge the dataframes on the "State" column
merged_df = energy.merge(ippu, on="State", how="outer", suffixes=('_energy', '_ippu'))
merged_df = merged_df.merge(waste, on="State", how="outer")
merged_df = merged_df.merge(afolu, on="State", how="outer")

# Rename columns for clarity
merged_df.columns = ['State', 'energy_values', 'ippu_values', 'waste_values', 'afolu_values']

# Replace NaN values with 0 (in case some states are missing in certain files)
merged_df.fillna(0, inplace=True)

# Sum the values for each state across all categories
merged_df['total_values'] = merged_df[['energy_values', 'ippu_values', 'waste_values', 'afolu_values']].sum(axis=1)

# Create a new DataFrame with only "State" and "total_values"
result_df = merged_df[['State', 'total_values']]

# Sort the values in descending order of 'total_values'
result_df = result_df.sort_values(by='total_values', ascending=False)

# Save the result to a new CSV file
result_df.to_csv('combined_states_2010.csv', index=False)

# Display a confirmation message
print("Data combined, sorted, and saved to 'combined_states_2010.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('combined_states_2010.csv')

# ---------------------- First Graph: Top 10 Emitting States with National Average Line ----------------------

# Sort the data by total_values in descending order and select the top 10 states
top_10_states = data.sort_values(by='total_values', ascending=False).head(10)

# Calculate the national average
national_average = data['total_values'].mean()

# Convert values to millions for plotting
top_10_states['total_values'] = top_10_states['total_values'] / 1e6
national_average_millions = national_average / 1e6

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_10_states['State'], top_10_states['total_values'], color='skyblue', edgecolor='black')

# Add a horizontal line for the national average in millions
plt.axhline(national_average_millions, color='red', linestyle='--', linewidth=1.5, 
            label=f'National Average ({national_average_millions:.2f}M)')

# Adding labels and title
plt.title('Top 10 Greenhouse Gas Emitting States in 2010', fontsize=14, weight='bold')
plt.xlabel('State', fontsize=12)
plt.ylabel('Total Emissions (in millions of tonnes)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Add a legend for the national average line
plt.legend(fontsize=10, loc='upper right')

# Display the graph
plt.tight_layout()
plt.show()
```
