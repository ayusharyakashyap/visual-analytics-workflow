```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("pie.csv")


countries = ["Zimbabwe", "South Africa", "Namibia"]
colors = ["red", "blue", "green", "gray"]

fig, axs = plt.subplots(1, len(df), figsize=(20, 5), subplot_kw=dict(aspect="equal"))


for i, row in df.iterrows():
    year = row["Year"]
    country_values = row[countries].tolist()
    rest_of_world = 100 - sum(country_values)
    values = country_values + [rest_of_world]  
    labels = countries + ["Rest of the World"]
    

    axs[i].pie(values, labels=None, colors=colors, startangle=90, autopct='%1.1f%%')
    axs[i].set_title(f"{int(year)}")

fig.legend(labels, loc="upper center", ncol=4, bbox_to_anchor=(0.5, 0.05))


plt.suptitle("Percentage Distribution for Selected Countries and Rest of the World (2004-2008)")
plt.tight_layout(rect=[0, 0.1, 1, 0.95])
plt.show()
```
