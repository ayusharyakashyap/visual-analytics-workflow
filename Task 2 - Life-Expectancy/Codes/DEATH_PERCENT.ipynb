```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Year": [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015],
    "Zimbabwe": [53, 52, 51, 50, 48, 46, 44, 40, 36.4, 32.6, 29.1, 25.6, 23.1, 21, 19.8, 18],
    "South Africa": [30, 34, 35, 37, 38, 37, 38.2, 37, 36.6, 34.9, 32.6, 29.8, 26.7, 24, 23, 21],
    "Namibia": [32, 31, 36, 38, 38, 36, 33.3, 29, 25.7, 24.4, 23.8, 22.6, 21.6, 21, 22, 22]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  
for i, col in enumerate(df.columns[1:]):
    plt.plot(df["Year"], df[col], marker='o', label=col, color=colors[i])

plt.title("Percentage of Deaths due to HIV/AIDS", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=14)
plt.legend(title="Countries", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
```

```python
plt.show()
```
