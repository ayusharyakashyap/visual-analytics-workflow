```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('TheContinent-life-exp-2010.csv') 


df.columns = df.columns.str.strip()

top_10 = df.nlargest(10, 'Life Expectancy')
bottom_10 = df.nsmallest(10, 'Life Expectancy')


continent_colors = {
    'Africa': 'orange',
    'Asia': 'lightblue',
    'Europe': 'lightgreen',
    'North America': 'yellow',
    'South America': 'pink',
    'Australia': 'purple',
    'Oceania': 'brown'
}


combined_df = pd.concat([top_10, bottom_10])


plt.figure(figsize=(12, 8))
sns.barplot(data=combined_df, x='Life Expectancy', y='Countries, territories and areas', 
            hue='Continent', palette=continent_colors)
plt.title('Top 10 and Bottom 10 Countries by Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Country')
plt.legend(title='Continent', loc='upper right')
plt.show()
```
