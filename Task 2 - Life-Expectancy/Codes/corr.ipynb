```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("correlation.csv")
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
df['rural_population_percentage'] = (df['rural_population'] / df['population']) * 100
```

```python
df_filtered = df[(df['date'].dt.year >= 2000) & (df['date'].dt.year <= 2015) & 
                 df['country'].isin(['South Africa', 'Zimbabwe', 'Zambia', 'Rwanda', 'Malawi', 'Algeria', 'Egypt'])]
```

```python
correlation_columns = ['birth_rate', 'death_rate', 'life_expectancy_at_birth', 
                        'intentional_homicides', 'HIV', 'health-exp', 'rural_population_percentage']
```

```python
df_corr = df_filtered[correlation_columns]


correlation_matrix = df_corr.corr()

life_expectancy_corr = correlation_matrix['life_expectancy_at_birth']


print("Correlations with Life Expectancy:\n")
print(life_expectancy_corr)


plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix ')
plt.show()
```
